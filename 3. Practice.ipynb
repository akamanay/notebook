## Joining tables

```python
# Join tables by Id

merged_data = pd.merge(data, customer_data_p, on='customer_id', how='outer', indicator=True)

#print(merged_data)
merged_data
```

```python
import pandas as pd
from functools import reduce

# Compile the list of DataFrames
data_frames_data = [data, customer_data_p, customer_data]
data_frames_order = [orders_data, customer_data_p, customer_data]

df_merged_data = reduce(lambda left, right: pd.merge(left, right, on='customer_id', how='inner'), data_frames_data)
df_merged_order = reduce(lambda left, right: pd.merge(left, right, on='customer_id', how='inner'), data_frames_order)

# Optionally, fill NaN values with a placeholder string
#df_merged_data = df_merged_data.fillna('void')
#df_merged_order = df_merged_order.fillna('void')

df_merged_data['customer_name'] = df_merged_data["customer_name_x"]
df_merged_order['customer_name'] = df_merged_data['customer_name_x']

df_merged_data = df_merged_data.drop(columns=['customer_name_x', 'customer_name_y'])
df_merged_order = df_merged_order.drop(columns=['customer_name_x', 'customer_name_y'])

data_frames_data = [df_merged_data, product_data, product_data_p]
data_frames_order = [df_merged_order, product_data, product_data_p]

df_merged_data = reduce(lambda left, right: pd.merge(left, right, on='product_id', how='inner'), data_frames_data)
df_merged_order = reduce(lambda left, right: pd.merge(left, right, on='product_id', how='inner'), data_frames_order)

df_merged_data['product_name'] = df_merged_data["product_name_x"]
df_merged_order['product_name'] = df_merged_data['product_name_x']

df_merged_data = df_merged_data.drop(columns=['product_name_x', 'product_name_y'])
df_merged_order = df_merged_order.drop(columns=['product_name_x', 'product_name_y'])

df_merged_data = df_merged_data.drop(columns=['product_id', 'customer_id'])
df_merged_order = df_merged_order.drop(columns=['product_id', 'customer_id'])

result = pd.concat([df_merged_order, df_merged_data], keys=['df_merged_order', 'df_merged_data'], join='outer', ignore_index=True)
result = result.sort_values(by='order_id')
```
